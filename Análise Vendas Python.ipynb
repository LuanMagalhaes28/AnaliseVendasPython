```python
#REMOVER OS ESPAÇOS DOS NOMES DAS COLUNAS
df.columns = df.columns.str.strip()
df.columns
```

```python
#Total Faturado e Qtde Total vendida
df['Valor Final'].sum() , df['Quantidade'].sum()
```

```python
#Total faturado por Loja
df[['ID Loja', 'Valor Final']].groupby('ID Loja').sum().sort_values(by='Valor Final', ascending=False)
```

```python
#Qtde Total Vendida por Loja
df[['ID Loja', 'Quantidade']].groupby('ID Loja').sum().sort_values(by='Quantidade', ascending= False)
```

```python
#ticket medio por loja
#to_frame = transformar em tabela
tk_f = (df[['ID Loja', 'Valor Final']].groupby('ID Loja').sum()['Valor Final'] / df[['ID Loja', 'Quantidade']].groupby('ID Loja').sum()['Quantidade']).to_frame(name='Ticket Médio')
tk_f.sort_values(by='Ticket Médio', ascending=False)
```

```python
#TOP 05 Produtos mais vendidos
df[['Produto', 'Quantidade']].groupby('Produto').sum().sort_values(by='Quantidade', ascending=False).head(5)
```

```python
#TOP 05 Produtos mais Faturados
df[['Produto', 'Valor Final']].groupby('Produto').sum().sort_values(by= 'Valor Final', ascending= False).head(5)
```

```python
#ticket medio por produto
tk_p = (df[['Produto', 'Valor Final']].groupby('Produto').sum()['Valor Final']/df[['Produto', 'Quantidade']].groupby('Produto').sum()['Quantidade']).to_frame(name='Tk_Medio')
tk_p.sort_values(by= 'Tk_Medio', ascending= False)
```

```python
#Correlação entre Qtde e Vl. Unit
df[['Quantidade', 'Valor Unitário']].corr()
#Correlação praticamente nula - Produtos de baixo valor com alta demanda.
```
